# **Preprocessing**

```python
import re

def replce_character(text):
    new_text = text.replace('-', ',')
    return new_text

def replace_city(text):
    # ='tphcm' -> 'thành phố hcm'
    text = re.sub(r'\btphcm\b', 'thành phố hcm', text, flags=re.IGNORECASE)
    
    # 'tphn' -> 'thành phố hn'
    text = re.sub(r'\btphn\b', 'thành phố hn', text, flags=re.IGNORECASE)
    
    # 'tp', 'tp.', 'tp .' -> 'thành phố'
    text = re.sub(r'\btp\.?\s?([^\W\d_]*)\b', r'thành phố \1', text, flags=re.IGNORECASE)
    
    # 'hcm' -> hồ chí minh
    text = re.sub(r'\bhcm\b', 'hồ chí minh', text, flags=re.IGNORECASE)
    
    # 'hn' -> hà nội
    text = re.sub(r'\bhn\b', 'hà nội', text, flags=re.IGNORECASE)
    
    return text

def replace_district(text):
    #' q<number>' thành 'quận <number>' (ex: q1 -> quận 1)
    text = re.sub(r'\bq(\d+)\b', r'quận \1', text, flags=re.IGNORECASE)
    
    #= p<number>' -> 'phường <number>' (ex: p5 -> phường 5)
    text = re.sub(r'\bp(\d+)\b', r'phường \1', text, flags=re.IGNORECASE)
    
    # 'p.', 'p .', 'p . <text>' -> 'phường'
    text = re.sub(r'\bp\.?\s?\b', 'phường ', text, flags=re.IGNORECASE)
    
    # Thay thế 'q.', 'q .', 'q . <text>' -> 'quận'
    text = re.sub(r'\bq\.?\s?\b', 'quận ', text, flags=re.IGNORECASE)
    
    return text

def format_text(text):
    # Remove spaces before and after commas and periods
    text = re.sub(r'\s*([,.])\s*', r'\1', text)
    
    # Remove double spaces
    text = re.sub(r'\s+', ' ', text)
    
    # Replace dot
    text = text.replace('.', ' ')
    
    return text

def covert_town(text):
    # 'tx' -> 'thị xã'
    text = text.replace('tx', 'thị xã')
    return text

def preprocessing_text(text):
    text = replce_character(text)
    text = replace_city(text)
    text = replace_district(text)
    text = format_text(text)
    text = covert_town(text)
    return text

# Example:
# Input: "q1, p5, tphcm, tx an khê"
# Output: "quận 1, phường 5, thành phố hồ chí minh, thị xã an khê"
```

```python
def add_prefix_to_location(address):
    # Provinces list
    provinces = ["long an", "quảng nam", "quảng ngãi", "quảng ninh", "bình dương", "bình phước", "bình thuận", 
                 "đồng nai", "đồng tháp", "an giang", "bà rịa - vũng tàu", "bạc liêu", "bắc giang", "bắc kạn", 
                 "bắc ninh", "bến tre", "cà mau", "cao bằng", "đắk lắk", "đắk nông", "điện biên", "gia lai", 
                 "hà giang", "hà nam", "hà tĩnh", "hải dương", "hậu giang", "hòa bình", "hưng yên", "khánh hòa", 
                 "kiên giang", "kon tum", "lai châu", "lâm đồng", "lạng sơn", "lào cai", "nam định", "nghệ an", 
                 "ninh bình", "ninh thuận", "phú thọ", "phú yên", "quảng bình", "sóc trăng", "sơn la", "tây ninh", 
                 "thái bình", "thái nguyên", "thanh hóa", "thừa thiên huế", "tiền giang", "trà vinh", "tuyên quang", 
                 "vĩnh long", "vĩnh phúc", "yên bái"]
    
    # Central cities list 
    cities = ["hà nội", "hồ chí minh", "đà nẵng", "hải phòng", "cần thơ"]

    # Keywords
    keywords = {
        'đường': ['đường', 'đg', 'dg'],
        'phường': ['phường', 'p.'],
        'xã': ['xã', 'x.'],
        'thị xã': ['thị xã', 'tx'],
        'quận': ['quận', 'q.', 'quận'],
        'huyện': ['huyện', 'h.'],
        'tỉnh': ['tỉnh'],
        'thành phố': ['thành phố', 'tp.', 'tp']
    }

    def add_prefix(segment, prefix):
        if not any(word in segment.lower() for word in keywords[prefix]):
            segment = f"{prefix} {segment.strip()}"
        return segment

    def classify_address(address):
        segments = [seg.strip() for seg in address.split(',')]
        length = len(segments)

        if length > 0:
            # Check for city/province
            for i in range(length - 1, -1, -1):
                seg = segments[i].lower()
                if any(city in seg for city in cities):
                    segments[i] = add_prefix(segments[i], 'thành phố')
                    if i - 1 >= 0:
                        segments[i - 1] = add_prefix(segments[i - 1], 'quận')
                    if i - 2 >= 0:
                        segments[i - 2] = add_prefix(segments[i - 2], 'phường')
                    if i - 3 >= 0:
                        segments[i - 3] = add_prefix(segments[i - 3], 'đường')
                    break
                elif any(province in seg for province in provinces):
                    segments[i] = add_prefix(segments[i], 'tỉnh')
                    if i - 1 >= 0:
                        segments[i - 1] = add_prefix(segments[i - 1], 'huyện')
                    if i - 2 >= 0:
                        segments[i - 2] = add_prefix(segments[i - 2], 'xã')
                    if i - 3 >= 0:
                        segments[i - 3] = add_prefix(segments[i - 3], 'đường')
                    break

        return ', '.join(segments)

    return classify_address(address)

# Example:
# Input: d2, bình thạnh, hcm
# Output: đường d2, quận bình thạnh, thành phố hcm
```

```python
def urban_districts(sentence):
    
    urban_districts = [
    "Bình Thạnh", "Gò Vấp", "Thủ Đức", "Tân Bình", "Tân Phú","Bình Tân",
    "Tân Bình", "Ba Đình", "Hoàn Kiếm", "Tây Hồ", "Long Biên", "Cầu Giấy",
    "Hai Bà Trưng", "Đống Đa", "Hoàng Mai", "Thanh Xuân", "Hà Đông",
    "Nam Từ Liêm", "Bắc Từ Liêm", "Gia Lâm", "Sóc Sơn", "Đông Anh",
    "Mê Linh", "Thường Tín", "Hoài Đức", "Quốc Oai","Thanh Trì"]
    
    # District names pattern
    district_pattern = r'\b(quận\s*)?\b(' + '|'.join(map(re.escape, urban_districts)) + r')\b'

    # Replace district names without "quận" with "quận" + name
    def replace(match):
        district_name = match.group(2)
        if not match.group(1):  # Nếu không có "quận" ở đầu
            return 'quận ' + district_name
        else:
            return 'quận ' + district_name  # Giữ nguyên "quận" nếu đã có

    # Replace district names (ignore case)
    new_sentence = re.sub(district_pattern, replace, sentence, flags=re.IGNORECASE)
    return new_sentence

def rural_districts(sentence):
    
    rural_districts = [
    "Bình Chánh", "Cần Giờ", "Củ Chi", "Hóc Môn", "Nhà Bè",
    "Đan Phượng", "Thạch Thất", "Chương Mỹ", "Phúc Thọ", "Mỹ Đức",
    "Ứng Hòa", "Thanh Oai", "Sơn Tây", "Ba Vì"]
    
    # District names pattern
    district_pattern = r'\b(huyện\s*)?\b(' + '|'.join(map(re.escape, rural_districts)) + r')\b'

    # Replace district names without "huyện" with "huyện" + name
    def replace(match):
        district_name = match.group(2)
        if not match.group(1):  # Nếu không có "huyện" ở đầu
            return 'huyện ' + district_name
        else:
            return 'huyện ' + district_name  # Giữ nguyên "huyện" nếu đã có

    # Replace district names (ignore case)
    new_sentence = re.sub(district_pattern, replace, sentence, flags=re.IGNORECASE)
    return new_sentence

def urban_or_rural(sentence):
    # Apply urban district replacements
    sentence = urban_districts(sentence)
    
    # Apply rural district replacements
    sentence = rural_districts(sentence)
    
    return sentence

# Example:
# Input: bình thạnh, gò vấp, quận thủ đức, nhà bè, mỹ đức, huyện ba vì
# Output: quận bình thạnh, quận gò vấp, quận thủ đức, huyện nhà bè, huyện mỹ đức, huyện ba vì
```

```python
def add_street_to_address(text):
    # Định nghĩa biểu thức chính quy để tìm các cụm từ dạng số nhà theo sau bởi tên đường
    pattern = re.compile(r'(\d+)\s*(\/\s*\d+)*\s+([^\d]+)')
    
    # Thay thế các cụm từ tìm được bằng cách chèn thêm từ 'đường'
    modified_text = pattern.sub(r'\1\2 đường \3', text)
    
    return modified_text

# Example:
# Input: 179 / 68 hai bà trưng
# Output: 179 / 68 đường hai bà trưng
```

# **Extract location phrases from a given text based on named entities**


```python
def extract_location_phrases(text, entities):
    keywords = ['đường', 'xã', 'phường', 'huyện', 'tỉnh', 'thành phố']
    if len(entities) == 1 or not any(keyword in text.lower() for keyword in keywords):
        return [text]
    
    location_phrases = []
    current_phrase_start = None
    current_phrase_end = None
    
    for i in range(len(entities)):
        if entities[i]['entity'] == 'B-LOCATION':
            if current_phrase_start is not None and current_phrase_end is not None:
                location_phrases.append(text[current_phrase_start:current_phrase_end])
            current_phrase_start = entities[i]['start']
            current_phrase_end = entities[i]['end']
        else:
            if current_phrase_start is not None:
                current_phrase_end = entities[i]['end']
        
        if i == len(entities) - 1:
            if current_phrase_start is not None:
                location_phrases.append(text[current_phrase_start:current_phrase_end])
    
    return location_phrases

# Example:
# Input: "Chung cư K26, đường 42, Phường Bình Trưng Tây, Quận 2"
# Output: ["Chung cư K26", "đường 42", "Phường Bình Trưng Tây", "Quận 2"]
```

```python
segs = []
for add in X:
    segs.append({add:extract_location_phrases(add, nlp(add))})
```

# **Classifying the components of addresses from a list of addresses**

```python
def classify_address(address_list):
    provinces_cities = ["hà nội", "hồ chí minh", "đà nẵng", "hải phòng", "cần thơ", "long an", "quảng nam", 
                        "quảng ngãi", "quảng ninh", "bình dương", "bình phước", "bình thuận", "đồng nai", 
                        "đồng tháp", "an giang", "bà rịa - vũng tàu", "bạc liêu", "bắc giang", "bắc kạn", 
                        "bắc ninh", "bến tre", "cà mau", "cao bằng", "đắk lắk", "đắk nông", "điện biên", 
                        "gia lai", "hà giang", "hà nam", "hà tĩnh", "hải dương", "hậu giang", "hòa bình", 
                        "hưng yên", "khánh hòa", "kiên giang", "kon tum", "lai châu", "lâm đồng", "lạng sơn", 
                        "lào cai", "nam định", "nghệ an", "ninh bình", "ninh thuận", "phú thọ", "phú yên", 
                        "quảng bình", "sóc trăng", "sơn la", "tây ninh", "thái bình", "thái nguyên", 
                        "thanh hóa", "thừa thiên huế", "tiền giang", "trà vinh", "tuyên quang", "vĩnh long", 
                        "vĩnh phúc", "yên bái"]

    keywords = {
        'đường': ['đường', 'đg', 'dg'],
        'phường': ['phường', 'p.'],
        'xã': ['xã', 'x.'],
        'thị xã': ['thị xã'],
        'quận': ['quận', 'q.', 'quận'],
        'huyện': ['huyện', 'h.'],
        'tỉnh': ['tỉnh'],
        'thành phố': ['thành phố', 'tp.', 'tp']
    }

    def classify_token(token):
        token = token.strip().lower()
        for key, value in keywords.items():
            for v in value:
                if v in token:
                    return key
        for province in provinces_cities:
            if province in token:
                if token == 'hà nội' or token == 'hồ chí minh':
                    return 'thành phố'
                else:
                    return 'tỉnh'
        return None

    def extract_classified_tokens(tokens):
        classified = {
            'duong': None,
            'px_cat': None,
            'px': None,
            'qh_cat': None,
            'qh': None,
            'tinh_cat': None,
            'tinh': None,
            'Address_ch': []
        }
        
        for token in tokens:
            classification = classify_token(token)
            if classification:
                if classification == 'đường':
                    classified['duong'] = token
                elif classification in ['phường', 'xã']:
                    classified['px_cat'] = classification
                    classified['px'] = token.replace(classification, '').strip()
                elif classification in ['quận', 'huyện']:
                    classified['qh_cat'] = classification
                    classified['qh'] = token.replace(classification, '').strip()
                elif classification in ['tỉnh', 'thành phố']:
                    classified['tinh_cat'] = classification
                    classified['tinh'] = token.replace(classification, '').strip()
            else:
                classified['Address_ch'].append(token)
        
        classified['Address_ch'] = ' '.join(classified['Address_ch'])
        return classified

    classified_addresses = []
    for item in address_list:
        for addr, tokens in item.items():
            classified = extract_classified_tokens(tokens)
            classified['Address'] = addr
            # Thêm số nhà vào cột Address_ch
            classified['Address_ch'] = add_street_to_address(classified['Address_ch'])
            classified_addresses.append(classified)
    
    # Create a DataFrame from the classified addresses with columns in reverse order
    df = pd.DataFrame(classified_addresses, columns=[
        'Address', 'tinh', 'tinh_cat', 'qh', 'qh_cat', 'px', 'px_cat', 'duong', 'Address_ch'
    ])
    return df

df1 = classify_address(segs)
df1
```
